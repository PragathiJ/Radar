```python
def Hue_matting(*argv):

    while(TRUE):
        #print(argv[0].shape)
        ret, frame =  argv[0].read()
        if(ret == False):
            print("I am here")
            break
        else:
            #frame = (frame/255.0).astype(np.float32)
            hsv = cv.cvtColor(frame, cv.COLOR_BGR2HSV)
            h,s,v = cv.split(frame);

            u_green = np.array([np.max(h), np.max(s), np.max(v)])
            l_green = np.array([np.average(h)/2, np.average(s)/2, np.average(v)/2]).astype('uint8')

            mask = cv.inRange(hsv, l_green, u_green)
            res = cv.bitwise_and(frame, frame, mask = mask)

            #ret,alpha = cv.threshold(f, 1, 1, cv.THRESH_BINARY);
            #ret,alpha = cv.threshold(-1*alpha, 0, 0, cv.THRESH_TRUNC);

            f = frame - res
            f = np.where(f == 0, argv[1], f)
            #f = cv.cvtColor(f, cv.COLOR_HSV2BGR)

            cv.imshow("Hue", f)
            cv.waitKey(10)

    #inputVideo.release()
    #cv.destroyAllWindows()
```

```python
def difference_matting(*argv):
    img = argv[2].astype('float32') / 255.0

    while(TRUE):
        ret, frame =  argv[0].read()
        if(ret == False):
            print("I am here")
            break
        else:

            frame = (frame/255.0).astype(np.float32)
            #frame = cv.GaussianBlur(frame, (5, 5), 0)
            background = (np.zeros(frame.shape)).astype('float32')
            # print(background.shape)
            background[:,:,1] = (np.average(frame[:,:,1]))
            background[:,:,0] = (np.average(frame[:,:,0]))
            background[:,:,2] = (np.average(frame[:,:,2]))
            #print((frame[0,0,:]))
            #print(np.average(frame[:,0,1]))

            image = frame - background
            image = np.sqrt(np.square(image[:,:,0]) + np.square(image[:,:,1]) + np.square(image[:,:,2]))

            #print(np.max(image))
            alpha = np.where(image<((np.max(image)-np.min(image))/2.5), 0.0, 1.0)

            #print(alpha)
            #cv.imshow("res", image)
            #cv.waitKey(10)

            video_channels = cv.split(frame)
            image_channels = cv.split(img)


            for i in range(3):
                ret = cv.multiply(alpha.astype(np.float32),video_channels[i],video_channels[i])
                ret = cv.multiply(argv[1] - alpha.astype(np.float32), image_channels[i],image_channels[i])

            #do not use numpy multiply
            #print(frame.shape)
            #print(img.shape)
            foreground = cv.merge(video_channels)
            background = cv.merge(image_channels)

            res_diff = foreground + background
            #print(res.shape)
            #res = np.int8(res)*255
            #res = np.clip(res, 0, 255) 
            #res = np.uint8(res)
            #print(np.average(frame[:,:,1]))
            #print(np.average(foreground[:,:,1]))
            #print(np.average(img[:,:,0]))

            cv.imshow("difference_matting", res_diff)
            cv.waitKey(10)

    #inputVideo.release()
    #cv.destroyAllWindows()
```
